# Рекурсивные функции

Рекурсивная функция — это та, которая вызывает сама себя.

Например, вспомним как высчитывается факториал: \
!7 = 7 * 6 * 5 * 4 * 3 * 2 * 1 или \
!7 = 7 * 6 * 5 * 4 * 3 * !2 или \
!7 = 7 * 6 * 5 * 4 * !3 или \
!7 = 7 * 6 * 5 * !4 или \
!7 = 7 * 6 * !5 или \
!7 = 7 * !6  

Поэтому в функции расчета факториала мы можем вызвать функцию факториал, но с n-1

````python
def factorial(n):
    if n == 1: # Это условие выхода из рекурсии. Когда ты пишешь рекурсивную фунцию, оно ВСЕГДА должно быть прописано
        return n
    else:
        return n * factorial(n - 1)
````
